```python
import matplotlib.pyplot as plt 
import random
import numpy as np
import math
```

# 1. Create Entities

```python
class Point:
    def __init__(self,x,y,color):
        self.__x=x
        self.__y=y
        self.__color=color
    
    @property
    def color(self):
        return self.__color
    
    @property
    def x(self):
        return self.__x

    @property
    def y(self):
        return self.__y
        
        
    
```

```python
unghi = random.uniform(-math.pi/2, math.pi/2)
line = [math.tan(unghi), random.uniform(-10,10)]
```

```python
def createPoints(n):
    points = []
    for _ in range(n):
        x = random.uniform(-10,10)
        y = random.uniform(-10,10)
        c = 'blue' if line[0] * x + line[1] > y else 'red'
        points.append(Point(x,y,c))
    return points
```

```python
points = createPoints(100)
```

```python
fig, ax = plt.subplots()

blue_x = [p.x for p in points if p.color == 'blue']
blue_y = [p.y for p in points if p.color == 'blue']

red_x = [p.x for p in points if p.color == 'red']
red_y = [p.y for p in points if p.color == 'red']

ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)

ax.scatter(red_x, red_y, color='red')
ax.scatter(blue_x, blue_y, color='blue')

x = np.linspace(-10,10, 1000)
y = line[0] * x + line[1]
ax.plot(x, y)
```
